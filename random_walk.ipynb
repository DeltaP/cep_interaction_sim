1. Fix distance metric to take into account periodic boundary conditions
1. Need to incorporate particle lifetimes (exponential decay)
2. Need to incorporate rate of new particles being formed
2. We only care about the minimum distance for an individual particle
3. New particle can not form at a site with an existing particle
3. N will change constantly
4. In the physical process when two particles are close enoug to activate they "dissapear"

```python
import collections
import random
import numpy as np
from scipy.spatial import distance

X = 99
Y = 99
Z = 99
N = 10
```

```python
def init():
    x = np.random.randint(X, size=(N,1))
    y = np.random.randint(Y, size=(N,1))
    z = np.random.randint(Z, size=(N,1))
    p = np.concatenate((x,y,z), axis=1)
    return p
```

```python
def min_distance(p):
    d = distance.cdist(p, p, 'euclidean')
    np.fill_diagonal(d,np.inf)
    m_d = np.amin(d,axis=1)
    return m_d
```

```python
def update(p):
    mx = np.full((N,1), X+1, dtype=np.int8)
    my = np.full((N,1), Y+1, dtype=np.int8)
    mz = np.full((N,1), Z+1, dtype=np.int8)
    m = np.concatenate((mx,my,mz), axis=1)
    step = np.random.choice((-1,0,1),(N,3))
    r = (p + step) % m
    return r
```

```python
p = init()
data = []
for i in range(99):
    data.extend(list(min_distance(p)))
    p = update(p)
```
